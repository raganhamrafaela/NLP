**Finalmente, nesse sétimo bloco de códigos utiliza a biblioteca Matcher para buscar padrões específicos de entidades nomeadas na coluna 'transcription' do DataFrame selecionado e imprimir o resultado.**

*Rafa



1.   Importe a classe "Matcher" do módulo "spacy.matcher". Isso tornará disponível a classe "Matcher" para ser usada em nosso código.

2.   Defina um padrão para a correspondência do Matcher. O padrão é uma lista de dicionários que especifica os critérios de correspondência para cada token. Neste caso, o padrão tem três dicionários, cada um com um critério diferente. O primeiro dicionário especifica que o token deve ter uma entidade nomeada do tipo "CHEMICAL". O segundo dicionário especifica que o token deve parecer um número. O terceiro dicionário especifica que o token deve ser ASCII.

3.   Crie uma instância da classe "Matcher" do SpaCy. A variável "nlp.vocab" deve ser um objeto que contém informações de vocabulário, como tokens, etiquetas POS, dependências sintáticas, etc.

4.   Use o método "matcher.add()" para adicionar o padrão de correspondência ao matcher. O primeiro argumento é o nome da correspondência e o segundo argumento é uma lista de padrões a serem correspondidos. Neste caso, estamos adicionando um padrão que corresponde a uma entidade química seguida por um número e uma string ASCII.




````python
from spacy.matcher import Matcher

pattern = [{'ENT_TYPE':'CHEMICAL'}, {'LIKE_NUM': True}, {'IS_ASCII': True}]
matcher = Matcher(nlp.vocab)
matcher.add("DRUG_DOSE", [pattern])
````

*Rafa



1.   Inicie um loop "for" que itera sobre as transcrições armazenadas na coluna "transcription" do dataframe "med_transcript_small". Para cada transcrição, aplique a função "nlp()" do SpaCy para processar o texto e gerar um objeto "doc".

2.   Use o objeto "matcher" para buscar correspondências no objeto "doc". O método "matcher()" retorna uma lista de tuplas contendo o ID da correspondência, o início do span correspondente e o final do span correspondente.

3.   Inicie outro loop "for" que itera sobre as correspondências encontradas. Para cada correspondência, extraia o ID da correspondência, o início e o fim do span correspondente.

4.   Use o objeto "nlp.vocab.strings" para obter a representação em string do ID da correspondência.

5.   Use os valores de início e fim do span correspondente para extrair o texto correspondente da transcrição original.

6.   Imprima na tela a representação em string do ID da correspondência, o início e o fim do span correspondente e o texto correspondente.







````python
for transcription in med_transcript_small['transcription']:
    doc = nlp(transcription)
    matches = matcher(doc)
    for match_id, start, end in matches:
        string_id = nlp.vocab.strings[match_id]  # get string representation
        span = doc[start:end]  # the matched span
        print(string_id, start, end, span.text)
````
